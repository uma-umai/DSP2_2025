### エラーハンドリングの基礎

```python
x = 1
y = 0
```

```python
x / y   #0で割っているのでエラーが起きる
```

```python
try:
    result = x / y
except ZeroDivisionError as e:
    print("エラーが発生しました:", e)
    print("0で割ることはできません。結果をNoneに設定します。")
    result = None

print("結果", result)
```

### 複数のエラーが発生する可能性がある場合

```python
try:
    result = x / z
except ZeroDivisionError:
    result = None
    print("エラーが発生しました", e)
    print("0で割ることはできません。結果をNoneに設定します。")
except NameError:
    result = None
    print("エラーが発生しました", e)
    print("変数が定義されていません。結果をNoneに設定します。")

print("結果", result)
```

```python
z = 3
try:
    result = x / z
except ZeroDivisionError:
    result = None
    print("エラーが発生しました", e)
    print("0で割ることはできません。結果をNoneに設定します。")
except NameError:
    result = None
    print("エラーが発生しました", e)
    print("変数が定義されていません。結果をNoneに設定します。")
else:
    #tryブロック内にてエラーがない時に実行される
    print("エラーは発生しませんでした。計算結果を表示します。")
finally:
    #エラーの有無にかかわらず実行される
    print("処理が終了しました。")

print("結果", result)
```

### システム終了以外の全てのエラーをハンドリングしたい場合

```python
try:
    result = x / y
except Exception as e:
    result = None
    print("エラーが発生しました:", e)
    print("エラーオブジェクト", type(e))
    print("エラーが発生したため結果をNoneに設定します。")
else:
    print("エラーは発生しませんでした。計算結果を表示します。")
finally:
    print("処理が終了しました。")

print("結果", result)
```

### WEBスクレイピングでのエラー処理


```python
import requests
```

```python
url = 'https://www.musashi.ac.jp/anvalid-page'  # 存在しないURL
response = requests.get(url)
print("ステータスコード:", response.status_code)

# ステータスコードが200でない場合に例外を発生させる
response.raise_for_status()
```

```python
url = 'https://www.musashi.ac.jp/anvalid-page' # 存在しないURL

try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.HTTPError as e:
    print("HTTPエラーが発生しました:", e)
except requests.exceptions.RequestException as e:
    print("リクエストエラーが発生しました:", e)
else:
    print("Webページに正常にアクセスできました。")
finally:
    print("webスクレイピング処理が終了しました。")
```
